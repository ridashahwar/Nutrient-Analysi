# Take high dimensional dataset & apply PCA - specify no. of components & also variance to be retained

```python
import pandas as pd
```

```python
data = pd.read_csv('nndb_flat.csv')
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
import numpy as np
```

```python
x= data.drop(['ID','FoodGroup','ShortDescrip','Descrip','CommonName','MfgName','ScientificName',], axis=1)
```

```python
x
```

```python
x.isnull().sum()
```

```python
data['Fiber_g'].head(20)
```

```python
x['Fiber_g']=x['Fiber_g'].replace(0.0,x['Fiber_g'].mean())
```

```python
x.head()
```

```python
x['VitC_mg']=x['VitC_mg'].replace(0.0,x['VitC_mg'].mean())
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()
x_scaled = ss.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.decomposition import PCA
# specify number of PC
p = PCA(n_components= 2)
```

```python
p_comp = p.fit_transform(x_scaled)
```

```python
p_comp
```

```python
p.explained_variance_ratio_
```

```python
# variance to be retained
p2 = PCA(.98)
```

```python
p2.fit_transform(x_scaled)
```

```python
p2.n_components_
```

```python
p1 = PCA(n_components= 4)
```

```python
p_comp1 = p1.fit_transform(x_scaled)
p_comp1
```

```python
p1.explained_variance_ratio_
```

```python
p1.n_components_
```
